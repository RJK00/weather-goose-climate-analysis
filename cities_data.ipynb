```python
import os
import pandas as pd
import plotly.graph_objects as go
```

```python
path = os.path.join('resources','cities.csv')
cities_csv = pd.read_csv(path)

# create a dataframe for viz
cities_df = pd.DataFrame(cities_csv)
cities_df.head(5)
```

```python
# convert csv to html table and save
cities_csv.to_html("table.html", index=False)
```

```python
fig1 = go.Figure(
    data=go.Scatter(x=cities_df['Lat'],
                    y=cities_df['Max Temp'],
                    mode='markers',
                    marker=dict(
                        color='darkcyan',
                        line=dict(
                            color='cornsilk',
                            width=0.3
                        )
                    )
    ),
    layout=go.Layout(
        title='Maximum Temperature (F) vs Latitude',
        xaxis_title='Latitude',
        yaxis_title='Max Temperature (F)',
        paper_bgcolor='whitesmoke',
        plot_bgcolor='snow',
        width=700
       

    ))

fig1.show()
```

```python
fig2 = go.Figure(
    data=go.Scatter(x=cities_df['Lat'],
                    y=cities_df['Humidity'],
                    mode='markers',
                    marker=dict(
                        color='darkcyan',
                        line=dict(
                            color='cornsilk',
                            width=0.3
                        )
                    )
    ),
    layout=go.Layout(
        autosize=True,
        title='Humidity vs Latitude',
        xaxis_title='Latitude',
        yaxis_title='Humidity (%)',
        paper_bgcolor='whitesmoke',
        plot_bgcolor='snow',
        width=700
    ))

fig2.show()
```
